## Preprocessing the data

When I tried to pass in the text files directly into the ```pandas.read_csv()``` this is what happened.

````python
messy_dataframe=pandas.read_csv("gen7ou-0.txt")
messy_dataframe.head()
````

Obviously this dataframe is very messy and hard to work with so I decided to try to convert the text files into JSON. While trying to look for ways to convert the text files into JSON, I found a function made by [Ali Turfah](https://github.com/aturfah) for a [project he made analyzing Pokemon statistics](https://github.com/aturfah/smogon-usage-stats). Below is the function I used to help convert each file into JSON readable format.

````python
def parse_data(file_url):
    file_data = {}
    txt_data = urlopen(file_url).read().decode()

    data_arr = txt_data.split("\n")
    #Remove the header rows/not relevant rows
    to_delete = [0, 1]
    for i in range(len(data_arr)):
        if " + " in data_arr[i] or "| Rank" in data_arr[i]:
            to_delete.append(i)

    to_delete = list(reversed(to_delete))
    for delete_index in to_delete:
        del data_arr[delete_index]

    for row in data_arr:
        if row == "":
            continue

        datum = " " + row.strip() + " "
        datum = sub("[ ]{1,50}", " ", datum)
        datum_arr = datum.split(" | ")

        #Remove unneeded datapoints
        datum_arr = datum_arr[1:5] 
        rank = datum_arr[0]
        name = datum_arr[1]
        usage_pcnt = datum_arr[2]
        usage_count = datum_arr[3]
        file_data[name] = {}
        file_data[name]["rank"] = rank
        file_data[name]["usage"] = usage_pcnt
        file_data[name]['count'] = usage_count
    
    return file_data
````

I used this function to convert each of the text files into JSON file, and that made it so I was able to make better pandas dataframes than before. The process of creating the JSON files was straightforward. All I did was first pass in the text file into the function then I used the output of the function and passed into a standard function that converted txt files to JSON files.

## Tidying the Data

The prime focus of this tutorial to see the affect of a Pokemon's stat distribution on its usage in competitive battles. I will merge the Pokemon Stats dataset and the Pokemon Usage dataset. I will only take the columns for the stats of the Pokemon from the Pokemon Stats dataset and ignore the other columns like ability and type weaknesses. 

First I create temporary collumns filled with NP.nan elements in the Pokemon usage dataframe.

````python
placeholder=np.full(899, np.nan)
pokemon_usage_dataframe_1500["HP"]=placeholder
pokemon_usage_dataframe_1500["Attack"]=placeholder
pokemon_usage_dataframe_1500["Defense"]=placeholder
pokemon_usage_dataframe_1500["Special Attack"]=placeholder
pokemon_usage_dataframe_1500["Special Defense"]=placeholder
pokemon_usage_dataframe_1500["Speed"]=placeholder
pokemon_usage_dataframe_1500.head()
````

Next I will copy the values for each Pokemon's stats by iterating through the Pokemon Stats dataframe and put them into the columns I created.

````python
index = 0
for first_row in pokemon_usage_dataframe_1500.iterrows():
    for second_row in pokemon_stats_dataframe_1500.iterrows():
        if first_row[1]["Pokemon"]==second_row[1]["Name"]:
            pokemon_usage_dataframe_1500["HP"][index] = second_row[1]["HP"]
            pokemon_usage_dataframe_1500["Attack"][index] = second_row[1]["Attack"]
            pokemon_usage_dataframe_1500["Defense"][index] = second_row[1]["Defense"]
            pokemon_usage_dataframe_1500["Special Attack"][index] = second_row[1]["Sp_Attack"]
            pokemon_usage_dataframe_1500["Special Defense"][index] = second_row[1]["Sp_Defense"]
            pokemon_usage_dataframe_1500["Speed"][index] = second_row[1]["Speed"]
    index=index+1
````

````python
pokemon_usage_dataframe_1500.head()
````

Now I will strip the percentage from the Usage and convert the values in the Usage column into floating point values. This will make it much more easier to use the Usage column when I am performining analysis on this data.

````python
for i in range(0,899):
    pokemon_usage_dataframe_1500["Usage"][i]=float(pokemon_usage_dataframe_1500["Usage"][i].strip('%'))/100
````

````python
pokemon_usage_dataframe_1500.head()
````

Now I will create a Heat Map based off the top 53 Pokemon used in Generation 7 Overused and see how the usage of each of these Pokemon increases as Elo increases. This list of Pokemon can be found here: https://www.smogon.com/dex/sm/formats/ou/

````python
Top_52_Pokemon =["Alakazam-Mega", "Azumarill", "Blacephalon", "Celesteela", "Chansey", "Charizard-Mega-X", "Charizard-Mega-Y", "Clefable", "Diancie-Mega", "Excadrill", "Ferrothorn", "Garchomp", "Garchomp-Mega", "Gliscor", "Greninja", "Greninja-Ash", "Gyarados-Mega", "Hawlucha", "Heatran", "Jirachi", "Kartana", "Keldeo", "Kommo-o", "Kyurem-Black", "Landorus-Therian", "Latias-Mega", "Lopunny-Mega", "Magearna", "Magnezone", "Manaphy", "Mawile-Mega", "Medicham-Mega", "Pelipper", "Rotom-Wash", "Sableye-Mega", "Scizor-Mega", "Serperior", "Skarmory", "Swampert-Mega", "Tangrowth", "Tapu Bulu", "Tapu Fini", "Tapu Koko", "Tapu Lele", "Tornadus-Therian", "Toxapex", "Tyranitar", "Tyranitar-Mega", "Victini", "Volcarona", "Zapdos"]
````

Here I initialize a new dataframe starting with a column containing the names of the Pokemon in the list.

````python
heat_df = pandas.DataFrame(Top_52_Pokemon, columns = ['Pokemon']) 
heat_df.head()
````

Now I will use the list of Pokemon in the Top 52 and find the usage stats from each tier corresponding to the Pokemon.

````python
usage_from_0=[]
usage_from_1500=[]
usage_from_1695=[]
usage_from_1825=[]

for pokemon in Top_52_Pokemon:
    exists=False
    for i in range(0,899):
        if pokemon_usage_dataframe_0["Pokemon"][i] == pokemon:
            exists=True
            usage_from_0.append(pokemon_usage_dataframe_0["Usage"][i])
    if not exists:
        print(pokemon)

for pokemon in Top_52_Pokemon:
    exists=False
    for i in range(0,899):
        if pokemon_usage_dataframe_1500["Pokemon"][i] == pokemon:
            exists=True
            usage_from_1500.append(pokemon_usage_dataframe_1500["Usage"][i])
    if not exists:
        print(pokemon)

for pokemon in Top_52_Pokemon:
    exists=False
    for i in range(0,899):
        if pokemon_usage_dataframe_1695["Pokemon"][i] == pokemon:
            exists=True
            usage_from_1695.append(pokemon_usage_dataframe_1695["Usage"][i])
    if not exists:
        print(pokemon)

for pokemon in Top_52_Pokemon:
    exists=False
    for i in range(0,899):
        if pokemon_usage_dataframe_1825["Pokemon"][i] == pokemon:
            exists=True
            usage_from_1825.append(pokemon_usage_dataframe_1825["Usage"][i])
    if not exists:
        print(pokemon)
````

I add the columns created in the code above to the heat_df dataframe.

````python
heat_df["Usage_0"]=usage_from_0
heat_df["Usage_1500"]=usage_from_1500
heat_df["Usage_1695"]=usage_from_1695
heat_df["Usage_1825"]=usage_from_1825
````

````python
heat_df.head()
````

This dataframe looks well organized but it has to be further modified in oder for it to work well with creating Heat Map. I will create yet another dataframe using the one I created above to achieve this purpose.

````python
tier_0_lst=[]
tier_1500_lst=[]
tier_1695_lst=[]
tier_1825_lst=[]
for i in range(0,51):
    tier_0_lst.append("Usage_0")
    tier_1500_lst.append("Usage_1500")
    tier_1695_lst.append("Usage_1695")
    tier_1825_lst.append("Usage_1825")

total_tier_lst=[]
for element in tier_0_lst:
    total_tier_lst.append(element)
for element in  tier_1500_lst:
    total_tier_lst.append(element)   
for element in  tier_1695_lst:
    total_tier_lst.append(element)  
for element in  tier_1825_lst:
    total_tier_lst.append(element)   

pokemon_lst=[]
for pokemon in heat_df["Pokemon"]:
    pokemon_lst.append(pokemon)
for pokemon in heat_df["Pokemon"]:
    pokemon_lst.append(pokemon)
for pokemon in heat_df["Pokemon"]:
    pokemon_lst.append(pokemon)
for pokemon in heat_df["Pokemon"]:
    pokemon_lst.append(pokemon)

useage_lst=[]
for useage in heat_df["Usage_0"]:
    useage_lst.append(useage)
for useage in heat_df["Usage_1500"]:
    useage_lst.append(useage)
for useage in heat_df["Usage_1695"]:
    useage_lst.append(useage)
for useage in heat_df["Usage_1825"]:
    useage_lst.append(useage)
````

Using the lists created above here is the dataframe needed.

````python
heat_map_df= pandas.DataFrame(pokemon_lst, columns = ['Pokemon']) 
heat_map_df["Tier"]=total_tier_lst
heat_map_df["Useage"]=useage_lst
heat_map_df.head()
````

Using the dataframe created above, here is the heat map for the Usage of each of these Pokemon over across different levels of competitive play.

````python
cmap = sns.diverging_palette(220,10, as_cmap=True)
pivot = heat_map_df.pivot(index='Tier', columns="Pokemon", values='Useage')
f, ax = plt.subplots(figsize=(12,9))
plt.title("Useage " + ' Heatmap')
sns.heatmap(pivot, cmap=cmap, square=True, linewidths=.5, ax=ax)
````

From the Heat Map above it is clear that usage of Landorus-Therian and Magearna increased as the Elo of players increased. Although less apparent in the map due to how popular Landorus and Magearna are, Pokemon like Volcarona and Greninja saw an increase in usage as Elo increased, Tornadus-Therian and Chansey saw a slight decrease in usage as Elo increased. Pokemon like Garchomp-Mega, and Tyranitar-Mega remained relatively the same in usage across all Elos.
